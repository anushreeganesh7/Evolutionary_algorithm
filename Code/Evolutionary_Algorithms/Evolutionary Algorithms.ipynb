```python
def generate_bit_string(l):
    rand_bit = np.random.randint(2, size=l)
    return rand_bit

def mutation(parent, mu):
    xm = np.copy(parent)
    for i in range(len(xm)):
        if np.random.rand()<mu:
            xm[i] = 1 - xm[i]
    return xm

def fitness(parent):
    np.sum(parent)

def best_fitness(gen, l, mu):
    best_fitness = []
    x = generate_bit_string(l)
    current_fitness = fitness(x)
    
    for g in range(gen):
        xm = mutation(x, mu)
        mutated_fitness = evaluate_fitness(xm)
        
        if mutated_fitness >= current_fitness:
            x = xm
            current_fitness = mutated_fitness
        
        best_fitnesses.append(current_fitness)
    
    return best_fitnesses

l = 100
gen = 1500
mu = 1 / l

best_fitnesses_original = best_fitness(gen, l, mu)
best_fitnesses_modified = best_fitness(gen, l, mu)

plt.plot(range(gen), best_fitnesses_original, label='Original Algorithm')
plt.plot(range(gen), best_fitnesses_modified, label='Modified Algorithm')
plt.xlabel('Generations')
plt.ylabel('Best Fitness')
plt.title('Comparison of Original and Modified (1+1)-GA')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def initialize_population(length):
    return np.random.randint(2, size=length)

def mutate(individual, mutation_rate):
    mutated_individual = np.copy(individual)
    for i in range(len(mutated_individual)):
        if np.random.rand() < mutation_rate:
            mutated_individual[i] = 1 - mutated_individual[i]
    return mutated_individual

def evaluate_fitness(individual):
    return np.sum(individual)

def one_plus_one_GA(length, generations, mutation_rate):
    best_fitnesses = []
    
    # Step 1: Initialization
    current_individual = initialize_population(length)
    current_fitness = evaluate_fitness(current_individual)
    
    for gen in range(generations):
        # Step 2: Mutation
        mutated_individual = mutate(current_individual, mutation_rate)
        
        # Step 3: Selection
        mutated_fitness = evaluate_fitness(mutated_individual)
        
        if mutated_fitness >= current_fitness:
            current_individual = mutated_individual
            current_fitness = mutated_fitness
        
        # Step 4: Record best fitness
        best_fitnesses.append(current_fitness)
    
    return best_fitnesses

if __name__ == "__main__":
    length = 100
    generations = 1500
    mutation_rate = 1 / length
    
    best_fitnesses_original = one_plus_one_GA(length, generations, mutation_rate)
    best_fitnesses_modified = one_plus_one_GA(length, generations, mutation_rate)
    
    plt.plot(range(generations), best_fitnesses_original, label='Simple Algorithm')
    plt.plot(range(generations), best_fitnesses_modified, label='Modified Algorithm')
    plt.xlabel('Generations')
    plt.ylabel('Best Fitness')
    plt.title('Comparison of Simple and Modified (1+1)-GA')
    plt.legend()
    plt.show()
```
