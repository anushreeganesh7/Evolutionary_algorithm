```python
import random
import string
filepath = '/Users/anushree/Desktop/Natural Computing/3-letter-words.txt'
TARGET_WORD = "PIC"
MUTATION_RATE = 0.2
#TODO incorporate the feedback given by right letter but wrong position

# Load word list from file
def load_word_list(file_path):
    with open(file_path, 'r') as file:
        words = file.read().splitlines()
    return words

#get select random words (might change to generate random sequences)
def generate_population(word_list, population_size):
    return random.sample(word_list, population_size)

# fitness based on correct items (order, value)
#returns num of matching chars and the indices of positions
def calculate_fitness(target_word = TARGET_WORD, guess=""):
    num_matching_chars = sum(1 for expected, actual in zip(target_word, guess) if expected == actual)
    correct_positions = [i for i, (expected, actual) in enumerate(zip(target_word, guess)) if expected == actual]
    return num_matching_chars, correct_positions

# Select parents based on fitness
def select_parents(population, num_parents):
    parents = []
    for _ in range(num_parents):
        fitness_scores = [calculate_fitness(target_word = TARGET_WORD, guess = individual) for individual in population]
        max_fitness_index = fitness_scores.index(max(fitness_scores))
        parents.append(population[max_fitness_index])
    return parents

#how to do crossover in this context?
def crossover(parent1, parent2):
    crossover_point = random.randint(1, len(parent1) - 1)
    child1 = parent1[:crossover_point] + parent2[crossover_point:]
    child2 = parent2[:crossover_point] + parent1[crossover_point:]
    return child1, child2

# Mutate (=change letter) if wrong letter and if letter is not at correct position
def mutate(individual, mutation_rate = MUTATION_RATE, target_word=TARGET_WORD):
    mutated_individual = ''
    for i, char in enumerate(individual):
        if char != target_word[i] and random.random() < mutation_rate:
            mutated_individual += random.choice(string.ascii_lowercase)
        else:
            mutated_individual += char
    return mutated_individual

def run(word_list, target_word, population_size, mutation_rate, num_generations):
    population = generate_population(word_list, population_size)
    for generation in range(num_generations):
        parents = select_parents(population, 2)
        offspring = []
        while len(offspring) < population_size:
            parent1, parent2 = random.sample(parents, 2)
            child1, child2 = crossover(parent1, parent2)
            child1 = mutate(child1, mutation_rate, target_word)
            child2 = mutate(child2, mutation_rate, target_word)
            offspring.extend([child1, child2])
        population = offspring
        best_guess = max(population, key=lambda x: calculate_fitness(target_word, x))
        print(f"Generation {generation+1}, Best Guess: {best_guess}, Fitness: {calculate_fitness(target_word, best_guess)}")
        if best_guess == target_word:
            print("Target word founnd")
            break
        
word_list = load_word_list(file_path=filepath)
fit = calculate_fitness("halop", "fanap")
pop = generate_population(word_list, 5)
par = select_parents(pop, 2)
mut = mutate("tes", 0.8,target_word="bet")
print(pop)
print(par)
run(word_list, TARGET_WORD, population_size=10, mutation_rate=MUTATION_RATE, num_generations=20)


```
